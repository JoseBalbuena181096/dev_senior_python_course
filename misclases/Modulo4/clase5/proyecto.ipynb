```python
# 1. contador de plabras clave en los comentarios, excluyendo palabras cortas

palabras = []
for r in resenna:
    # palabras.extend(r["comentario"].lower().split())
    palabras.extend(p for p in r["comentario"].lower().split() if len(p) > 3)
print(palabras)
```

```python
conteo_palabras = collections.Counter(palabras)
print("Palabras mas frecuentes")
print(conteo_palabras.most_common(3))
```

```python
# 2. Agrupación de reseñas por puntaje
grupo_resennas =  collections.defaultdict(list)
for r in resenna:
    grupo_resennas[r['puntaje']].append(r['comentario'])
print("Reseñas agrupadas por puntajes: ")
for puntaje, comentarios in grupo_resennas.items():
    print(f"Puntaje {puntaje}:{comentarios}")

print(grupo_resennas)
```

```python
# Generador combinaciones de palabras clave
print("Combinación de plabras con comentarios")
combinaciones = itertools.combinations(conteo_palabras.keys(), 2)
print(list(combinaciones))
```

```python
# Genera respuestas actomatica

respuestas = ["Gracias por tu momentario!" , "Nos alegra que te aya gustado", "Esperamos verte denuevo"]
permutaciones_respuestas = itertools.permutations(respuestas, 2)
print("Respuestas automaticas")
for respuesta in permutaciones_respuestas:
    print(" - ".join(respuesta))
```

```python
# 5. Graficar de barras palabras frecuentes
import random

# Configuración de la figura
plt.figure(figsize=(8,5))

# Generar colores aleatorios
colores = ["#" + ''.join(random.choices('0123456789ABCDEF', k=6)) for _ in range(5)]

# Graficar las 5 palabras más frecuentes
palabras, frecuencias = zip(*conteo_palabras.most_common(5))
plt.bar(palabras, frecuencias, color=colores)

# Etiquetas y título
plt.xlabel("Palabras")
plt.ylabel("Frecuencia")
plt.title("Top 5 de palabras más frecuentes en reseñas")

# Mostrar gráfico
plt.show()
```

```python
# 6.Grafica de pastel de puntajes
puntajes = [r["puntaje"] for r in resenna]
conteo_puntajes = collections.Counter(puntajes)

plt.Figure(figsize=(12, 12))
plt.pie(conteo_puntajes.values(),  labels=conteo_puntajes.keys(), autopct='%1.1f%%', startangle=90)
plt.title("Distribución por reseñas")
plt.show()
```
